```python
import folium
import math
import random
import numpy as np
```

```python
x = (23.795717, 90.406010)
map = folium.Map(location = (23.795717, 90.406010), zoom_start = 20, tiles ="OpenStreetMap")

fg = folium.FeatureGroup(name = "markers")
z = []

for i in range(0,10):    
    y = [x[0]+((random.randint(-250,250))/1000000), x[1]+((random.randint(-250,250))/1000000)]
    y = float('%.6f'%(y[0])), float('%.6f'%(y[1]))
    z.append(y)
```

```python
print(z)
```

```python
for i in z:
    fg.add_child(folium.Marker(location = i))
```

```python
map.add_child(fg)
```

```python
map.save("coordinates.html")
```

```python
import ast
```

```python
#Converting strings to tuples
for i in range (0,10):
    df.iloc[0, i] = ast.literal_eval(df.iloc[0, i])
    df.iloc[1, i] = ast.literal_eval(df.iloc[1, i])
    df.iloc[2, i] = ast.literal_eval(df.iloc[2, i])
```

```python
# tuples to array/lists
for i in range (0,10):
    df.iloc[0,i] = list(df.iloc[0,i])
    df.iloc[1,i] = list(df.iloc[1,i])
    df.iloc[2,i] = list(df.iloc[2,i])
```

```python
#generating randomized values for road20
for l in range (3,24):
    z = []
    for i in range(0,10):
        x = []
        for j in df.iloc[0,i]:
            j = j+((random.randint(-250,250))/1000000)
            j = float('%.6f'%(j))
            x.append(j)
        z.append(x)

    z.append("road20")
    df.loc[l] = z
#     df.iloc[1,i] = x
#     df.loc[0,"k"] = "road20"

#road22
for l in range (24,45):
    z = []
    for i in range(0,10):
        x = []
        for j in df.iloc[1,i]:
            j = j+((random.randint(-250,250))/1000000)
            j = float('%.6f'%(j))
            x.append(j)
        z.append(x)

    z.append("road22")
    df.loc[l] = z

#road24    
for l in range (45,66):
    z = []
    for i in range(0,10):
        x = []
        for j in df.iloc[2,i]:
            j = j+((random.randint(-250,250))/1000000)
            j = float('%.6f'%(j))
            x.append(j)
        z.append(x)

    z.append("road24")
    df.loc[l] = z

    
```
